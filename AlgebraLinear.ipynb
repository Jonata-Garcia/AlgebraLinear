```python
import numpy as np

# Matriz A (3x2)
A = np.array([[1, 2], [3, 4], [5, 6]])

# Transposta de A
A_transposta = np.transpose(A)

print("Transposta de A:", A_transposta)
```

```python
import numpy as np

# Vetor v
v = np.array([3, 4, 5])

# Norma do vetor v
norm_v = np.linalg.norm(v)

print("Norma do vetor:", norm_v)
```

```python
import numpy as np

# Vetor v
v = np.array([2, 3, 4])

# Escalar k
k = 5

# Multiplicao escalar
v_scaled = k * v

print("Vetor escalado:", v_scaled)
```

```python
import numpy as np
import scipy.linalg

# Matriz A
A = np.array([[4, 3, 2],
              [2, 1, 3],
              [3, 2, 4]])

# Decomposio LU de A
P, L, U = scipy.linalg.lu(A)

print("Matriz P (permutao):", P)
print("Matriz L (lower triangular):", L)
print("Matriz U (upper triangular):", U)
```

```python
import numpy as np

# Vetores u e v
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

# Produto interno
dot_product = np.dot(u, v)

print("Produto interno:", dot_product)
```
